```python
from array import array
import os
import glob,math
import operator
import matplotlib.pylab as plt
```

```python
x =[]
y = []
file=open('event_fraction_1.txt', 'r')
for line in file:
    if 'h_' in line and not 'MET' in line and not '_0l_' in line:
        x.append(line.split('_')[2]+'_'+line.split('_')[4].split(':')[0])
        y.append(float(line.split()[-1]))
plt.bar(x, y, align='center')
plt.xticks(x, x)
plt.yscale('log', nonposy='clip')
plt.xticks(rotation=90)
plt.ylabel(r"Fraction of ttbar events")
plt.title(r"TTBar Decay Check ")
plt.subplots_adjust(bottom=0.15)
plt.ylim(10**-9, 10**-3)
plt.savefig('eventfraction_ttbar.pdf', bbox_inches='tight')
plt.show()
plt.clf()
plt.cla()
plt.close('all')
```
